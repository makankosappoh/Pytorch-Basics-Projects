```python
#create a new tensor for mean median, mode, sum, product, max, min, standard deviation, variance , argmax, argmin
c = torch.randint(size=(2, 5), low=0, high=10, dtype=torch.float32)
print(c)
print(c.mean())
print(c.mean(dim =0))
print(c.median())
print(c.mode())
print(c.sum())
print(c.prod())
print(c.max())
print(c.min())
print(c.std())
print(c.var())
print(c.argmax())
print(c.argmin())
```

```python
#matrix operations
d = torch.tensor([[1, 2], [3, 4]])
e = torch.tensor([[5, 6], [7, 8]])
print(d)
print(e)
print(d @ e) #matrix multiplication
print(d.mm(e)) #matrix multiplication
print(d * e) #elementwise multiplication
print(d * 10) #scalar multiplication
print(d.T) #transpose 
print(d.t()) #transpose
print(d.transpose(0, 1)) #transpose 0 and 1 dimensions
print(d.det()) #determinant
print(d.inverse()) #inverse
```

```python
#special functions
f = torch.randint(size=(2, 5), low=-10, high=10, dtype=torch.float32)
print(f)
print(torch.abs(f))
print(torch.sqrt(f)) #gives nan for negative values
print(torch.exp(f))
print(torch.log(f)) #gives nan for negative values and zero
print(torch.sin(f))
print(torch.cos(f))
print(torch.tan(f))
print(torch.sigmoid(f))
print(torch.relu(f,))
print(torch.softmax(f, dim=0))
print(torch.softmax(f, dim=1))
```

```python
#implace operations = operations that change the content of a tensor without changing its identity by adding _ to the end of the function name
g = torch.rand(2, 3)
print(g)
print(g.add(10)) #does not change g and g already eist in memeory taking its own space in memory and create new g tensor after adding 10 to g
print(g.add_(10)) #changes g permanently after adding 10 to g it will change g permanently and g will be stored in the same memory location as before but with new values after adding 10 to g
print(id(g)) #memory location of g
```

```python
#copying tensors wth issues
h = torch.rand(2, 3)
i = h
print(h)
print(i)
id(h), id(i) #h and i are pointing to the same memory location
```

```python
i[0][0] = 9
print(i)
print(h) #h also changes because i and h are pointing to the same memory location 
id(h), id(i) #h and i have the same memory location
```

```python
#copying tensors without issues
j = torch.rand(2, 3)
k = j.clone()
print(j)
print(k)
k[0][0] = 9
print(k)
print(j) #j does not change because k is a clone of j and k and j are separate tensors in memory
id(j), id(k) #j and k have different memory locations
```

```python
#cpu vs gpu
import torch, time

x = torch.randn(2000, 2000)
y = torch.randn(2000, 2000)

# CPU
t = time.time()
z = x @ y
print("CPU:", time.time() - t)

# GPU
if torch.cuda.is_available():
    x = x.cuda()
    y = y.cuda()
    torch.cuda.synchronize()

    t = time.time()
    z = x @ y
    torch.cuda.synchronize()
    print("GPU:", time.time() - t)
```
