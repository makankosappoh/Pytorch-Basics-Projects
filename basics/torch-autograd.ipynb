```python
#Autograd = automatic differentiation = computing gradients automatically
import torch
x = torch.tensor(2.0, requires_grad=True) #create a tensor with requires_grad=True to track gradients
y = x**2 + 3*x + 5 #define a function of x
print(x)
print(y)
```

```python
print(y.backward()) #compute the gradients meaning compute dy/dx
print(x.grad) #print the gradient of y with respect to x that is derived from the function y = x^2 + 3x + 5 is dy/dx = 2x + 3, so when x = 2, dy/dx = 2*2 + 3 = 7
```

```python
#y = x^2 calculating z = sin(y) and then calculating dz/dx
x = torch.tensor(2.0, requires_grad=True)
y = x**2
z = torch.sin(y)
print(z)
z.backward()
print(x.grad) #the gradient of z with respect to x is dz/dx = dz/dy * dy/dx = cos(y) * 2x, so when x = 2, y = 4, dz/dx = cos(4) * 2*2 = -1.5136
```

```python
#what if we y.grad in previous example
print(y.grad)
#y is not a scalar, so we need to specify the gradient of y with respect to itself, which is 1.0
```

```python
#example with cgpa = 6.7 and placement not placed ie 0 and we want to calculate the loss using binary cross entropy loss function and then calculate the gradients with respect weights and bias
x = torch.tensor(6.7) #input feature
y = torch.tensor(0.0) #target label

w = torch.tensor(1.0, requires_grad=True) #weights 
b = torch.tensor(0.0, requires_grad=True) #bias

z = w * x + b #linear regression model
print(z) 
y_pred = torch.sigmoid(z) #sigmoid activation function to get the predicted probability
print(y_pred)

loss = torch.nn.functional.binary_cross_entropy(y_pred, y)
print(loss)
loss.backward() #compute the gradients of loss with respect to weights and bias
print(w.grad) #print the gradient of loss with respect to weights   
print(b.grad) #print the gradient of loss with respect to bias
```

```python
#vector tensors and autograd
w = torch.tensor([1.0, 2.0, 3.0], requires_grad=True) #input feature vector
b = torch.tensor([0.0, 0.0, 0.0], requires_grad=True) #target labels 
z = w * x + b #linear regression model
print(z)
y_pred = torch.sigmoid(z) 
print(y_pred)
loss = torch.nn.functional.binary_cross_entropy(y_pred, y)
print(loss)
loss.backward() 
print(w.grad) 
print(b.grad)
```

```python
#sometime swe dont want to compute the gradients for some operations with 3 methods

# we can use torch.no_grad() context manager to disable gradient tracking for those operations
with torch.no_grad():
    w = torch.tensor(2.0, requires_grad=True)
    v = w**2
    print(v)


#we can also use detach() method to create a new tensor that shares the same data but does not require gradients
w = torch.tensor(2.0, requires_grad=True)
v = w**2
v.detach_()
print(v)

#we can also use requires_grad_() method to set requires_grad to False for those operations
w = torch.tensor(2.0, requires_grad=True)
v = w**2
v.requires_grad_(False)
print(v)

```
